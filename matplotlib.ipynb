```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#plotting the data
```

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title, x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine values')
plt.title('Sine wave')
plt.show()
```

```python
#plotting a parabola
x = np.linspace(-10,10,20)
y = x ** 2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x), 'g')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

```python
#BAR PLOTS
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
subject = ['Physics','M1','BXE','Chemistry','EM']
marks = [90,50,100,40,70]
ax.bar(subject,marks)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```

```python
#pie chart
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
ax.pie(marks,labels=subject, autopct = '%1.1f')
plt.show()
```

```python
#scatter plot
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
```

```python
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
# 3D scatter-plot
```

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,x,c = z, cmap='Blues')
plt.show()
```
